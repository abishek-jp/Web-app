```python
import pandas as pd
import numpy as np
X=pd.read_csv("/content/India Agriculture Crop Production(1).csv")
X.iloc[:20]
#introductory look at our dataset
```

```python
# Check for NaN values in each column and print column names
def check_nan_columns(df):
    """Check which columns contain NaN values and print their names"""
    nan_columns = df.columns[df.isna().any()].tolist()

    if nan_columns:
        print("Columns with NaN values:")
        for col in nan_columns:
            nan_count = df[col].isna().sum()
            print(f"  - {col}: {nan_count} NaN values")
    else:
        print("No columns contain NaN values")

    return nan_columns

# Example usage:
# Assuming you have a DataFrame called 'df'
# nan_cols = check_nan_columns(df)

# For your specific case with X_train_final and y_train:
print("Checking X_train_final:")
nan_cols_X = check_nan_columns(X)

```

```python
# Count NaN values per row
nan_per_row = X.isna().sum(axis=1)

# Find rows with more than 1 NaN
rows_with_multiple_nan = nan_per_row[nan_per_row > 1]

# Count such rows
count_multiple_nan_rows = len(rows_with_multiple_nan)

print(f"Number of rows with multiple NaN values: {count_multiple_nan_rows}")
print(f"Rows with multiple NaNs: {rows_with_multiple_nan}")

# Get the actual row indices
multiple_nan_indices = rows_with_multiple_nan.index.tolist()
print(f"Row indices with multiple NaNs: {multiple_nan_indices}")
```

```python
#imputation for "Crop"
# Step 1: district-level mode
district_mode = (
    X.groupby("District")["Crop"]
     .agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)

# Step 2: state-level mode
state_mode = (
    X.groupby("State")["Crop"]
     .agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)

# Step 3: function to fill with hierarchy
def fill_crop(row):
    if pd.isna(row["Crop"]):
        # try district mode
        district_val = district_mode.get(row["District"], np.nan)
        if pd.notna(district_val):
            return district_val
        # fallback to state mode
        state_val = state_mode.get(row["State"], np.nan)
        if pd.notna(state_val):
            return state_val
        # final fallback
        return "Unknown"
    else:
        return row["Crop"]

# Step 4: apply
X["Crop"] = X.apply(fill_crop, axis=1)
```

```python
#imputation for "Season"
# Step 1: calculate mode of Season for each Crop
crop_season_mode = (
    X.groupby("Crop")["Season"]
     .agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
)

# Step 2: function to fill missing Season
def fill_season(row):
    if pd.isna(row["Season"]):
        season_val = crop_season_mode.get(row["Crop"], np.nan)
        if pd.notna(season_val):
            return season_val
        return "Unknown"   # fallback if no mode for that crop
    else:
        return row["Season"]

# Step 3: apply
X["Season"] = X.apply(fill_season, axis=1)
```

```python
def impute_values(row):
    prod, area, yld = row["Production"], row["Area"], row["Yield"]

    # Case 1: Fill Yield
    if pd.isna(yld) and pd.notna(prod) and pd.notna(area) and area != 0:
        row["Yield"] = prod / area

    # Case 2: Fill Production
    elif pd.isna(prod) and pd.notna(yld) and pd.notna(area):
        row["Production"] = yld * area

    # Case 3: Fill Area
    elif pd.isna(area) and pd.notna(yld) and pd.notna(prod) and yld != 0:
        row["Area"] = prod / yld

    # Else â†’ do nothing, leave NaN
    return row

X = X.apply(impute_values, axis=1)
```

```python
# Check for NaN values in each column and print column names
def check_nan_columns(df):
    """Check which columns contain NaN values and print their names"""
    nan_columns = df.columns[df.isna().any()].tolist()

    if nan_columns:
        print("Columns with NaN values:")
        for col in nan_columns:
            nan_count = df[col].isna().sum()
            print(f"  - {col}: {nan_count} NaN values")
    else:
        print("No columns contain NaN values")

    return nan_columns

# Example usage:
# Assuming you have a DataFrame called 'df'
# nan_cols = check_nan_columns(df)

# For your specific case with X_train_final and y_train:
print("Checking X_train_final:")
nan_cols_X = check_nan_columns(X)
```

```python
mask_all_missing = X[["Production", "Area", "Yield"]].isna().all(axis=1)

count_all_missing = mask_all_missing.sum()
print(f"Number of rows where Production, Area, and Yield are all missing: {count_all_missing}")
#we will be dropping such rows
X = X.dropna(subset=["Production", "Area", "Yield"], how="all")
```

```python
# Check for NaN values in each column and print column names
def check_nan_columns(df):
    """Check which columns contain NaN values and print their names"""
    nan_columns = df.columns[df.isna().any()].tolist()

    if nan_columns:
        print("Columns with NaN values:")
        for col in nan_columns:
            nan_count = df[col].isna().sum()
            print(f"  - {col}: {nan_count} NaN values")
    else:
        print("No columns contain NaN values")

    return nan_columns

# Example usage:
# Assuming you have a DataFrame called 'df'
# nan_cols = check_nan_columns(df)

# For your specific case with X_train_final and y_train:
print("Checking X_train_final:")
nan_cols_X = check_nan_columns(X)
```

```python
indexes=X[X["Crop"].isin(["Other Kharif pulses","other oilseeds","Other Rabi pulses","Oilseeds total","Other Cereals","Other Summer Pulses"])].index
X_mod=X.drop(indexes)
rows, cols = X_mod.shape
print(f"Rows: {rows}, Columns: {cols}")
```

```python
pred=X_mod.Production
X_t=X_mod.drop(["Yield","Area Units","Production Units","Production","Year"],axis=1)
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X_t,pred,test_size=0.2)
```

```python
#main preprocessing
#categorical cols will be implemented using one hot encoding
List1=[]
for col in X_t.columns:
    if X_t[col].dtype=="object":
        List1.append(col)
#list 1 now contains all columns in the dataframe which have entries of object datatype in them i.e these are the columns containing categorical variables that are to be one hot encoded
X1=X_train.drop(List1,axis=1)
X2=X_valid.drop(List1,axis=1)
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder(sparse_output=False,handle_unknown="ignore")
P1=pd.DataFrame(OHE.fit_transform(X_train[List1]))
P2=pd.DataFrame(OHE.transform(X_valid[List1])) #using fit here causes data leakage as the model can now peek into validation data. so fit is not being used here.
P1.index=X_train.index
P2.index=X_valid.index
X_train_final=pd.concat([X1,P1],axis=1)
X_valid_final=pd.concat([X2,P2],axis=1)
X_train_final.head()
```

```python
#modelling using gradient boosting
from xgboost import XGBRegressor
model=XGBRegressor(n_estimators=1000,learning_rate=0.05,n_jobs=-1,random_state=0)
model.fit(X_train_final,y_train)
print(model.predict(X_valid_final))
model.score(X_valid_final,y_valid)
```

```python
model.score(X_train_final, y_train)
#Care has been taken to ensure that there is no validation data being leaked as mentioned in the previous steps
```

```python
#cross validation for a better picture of accuracy
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor

# Identify categorical columns (same as before)
categorical_cols = X_t.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X_t.select_dtypes(exclude=['object']).columns.tolist()

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ('num', 'passthrough', numerical_cols)
    ]
)

# Create full pipeline with preprocessing + model
full_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1, random_state=0))
])

# Now run cross-validation on the RAW data
cv_scores = cross_val_score(
    estimator=full_pipeline,  # Use the pipeline, not just the model
    X=X_t,                    # Raw data with categoricals
    y=pred,                   # Target
    cv=5,
    scoring='r2',
    n_jobs=-1
)

print("Cross-Validation RÂ² Scores:", cv_scores)
print(f"Mean CV RÂ²: {cv_scores.mean():.4f} (Â±{cv_scores.std()*2:.4f})")
```
